###  Часть 1. Описание набора данных и признаков

Предоставленный для анализа набор данных размером (84000, 11) имеет следующие признаки:                              
1) SeriousDlqin2yrs - испытает ли финансовые затруднения в течение следующих двух лет (1 - да, 0 - нет).            
2) RevolvingUtilizationOfUnsecuredLines - общий остаток по кредитным картам и личным кредитным линиям, за исключением недвижимости и задолженности в рассрочку, например, автокредитов, деленный на процентную долю кредитных лимитов.                                                                                                             
3) age - возраст заемщика.                                                                                                           
4) NumberOfTime30-59DaysPastDueNotWorse - количество раз, когда заемщик просрочил оплату на 30-59 дней.              
5) DebtRatio - ежемесячные выплаты по долгам, алименты, расходы на проживание, деленные на месячный процент валового дохода.                                                                      
6) MonthlyIncome - реальный ежемесячный доход.  
7) NumberOfOpenCreditLinesAndLoans)- количество открытых кредитов (в рассрочку, например, автокредит или ипотека) и кредитных линий (например, кредитных карт).  
8) NumberOfTimes90DaysLate - количество раз, когда заемщик просрочил оплату на 90 и более дней.  
9) NumberRealEstateLoansOrLines - количество ипотечных кредитов и кредитов на недвижимость, включая кредитные линии.  
10) NumberOfTime60-89DaysPastDueNotWorse - количество раз, когда заемщик просрочил оплату на 60-89 дней.  
11) NumberOfDependents - количество членов семьи, исключая себя (супруга, дети и т.д.).

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.pipeline import make_union, make_pipeline
```

```python
def num_data(data):
    data.NumberOfDependents.fillna(data.NumberOfDependents.mean(),inplace=True)
    
    return data.drop(columns=['MonthlyIncome', 'DebtRatio'])
def cat_data(data):
    data.MonthlyIncome.fillna(data.MonthlyIncome.mean(),inplace=True)
    
    data.loc[data['MonthlyIncome'] <= 5500, 'MonthlyIncome'] = 0,
    data.loc[(data['MonthlyIncome'] > 5500) & (data['MonthlyIncome'] <= 7000), 'MonthlyIncome'] = 1,
    data.loc[(data['MonthlyIncome'] > 7000) & (data['MonthlyIncome'] <= 22000), 'MonthlyIncome'] = 2,
    data.loc[(data['MonthlyIncome'] > 22000), 'MonthlyIncome'] = 3,

    data.loc[data['DebtRatio'] <= 500, 'DebtRatio'] = 0,
    data.loc[(data['DebtRatio'] > 500) & (data['DebtRatio'] <= 900), 'DebtRatio'] = 1,
    data.loc[(data['DebtRatio'] > 900), 'DebtRatio'] = 2,

    return data[['MonthlyIncome', 'DebtRatio']]
```

```python
pipe_un = make_union(*[make_pipeline(FunctionTransformer(num_data, validate=False), StandardScaler()),
                  make_pipeline(FunctionTransformer(cat_data, validate=False), OneHotEncoder(sparse=False))])
```

```python
new_data_X = pd.DataFrame(pipe_un.fit_transform(data_balanced.drop(columns='SeriousDlqin2yrs')))
new_data_y = data_balanced['SeriousDlqin2yrs']
new_data_X.head(15)
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, train_test_split
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_data_X, new_data_y, test_size=0.3)
```

Функция для построения roc curve

```python
def plot_roc(y_test,y_probs, auc, hld_y, pred_proba_holdout, auc_score_hld):
    fpr1, tpr1, thresholds1 = roc_curve(y_test, y_probs)
    fpr2, tpr2, thresholds2 = roc_curve(hld_y, pred_proba_holdout)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.plot(fpr1, tpr1, label='ROC curve (area = %0.2f)' % auc, marker='.')
    plt.plot(fpr2, tpr2, label='ROC curve holdout (area = %0.2f)' % auc_score_hld, marker='.')
    plt.title("ROC curve")
    plt.xlabel('false positive rate')
    plt.ylabel('true positive rate')
    plt.legend(loc="lower right")
    plt.show()
```

#### Логистическая регрессия

```python
param_grid = {
            'penalty' : ['l1','l2'], 
            'class_weight' : ['balanced', None], 
            'C' : [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
            }

grid_log = GridSearchCV(estimator = LogisticRegression(class_weight = 'balanced'), param_grid = param_grid , scoring = 'roc_auc')
grid_log.fit(X_train, y_train)
```

```python
%%time
pred_proba, pred_proba_holdout = grid_log.best_estimator_.predict_proba(X_test)[:, 1], grid_log.best_estimator_.predict_proba(hld_X)[:, 1]
auc_score = roc_auc_score(y_test, pred_proba)
auc_score_hld = roc_auc_score(hld_y, pred_proba_holdout)
print('roc auc score:', auc_score)
print('roc auc score holdout:', auc_score_hld)
```

###### roc curve 

```python
plot_roc(y_test, pred_proba, auc_score, hld_y, pred_proba_holdout, auc_score_hld)
```

###### изменение roc auc по C

```python
plt.plot(grid_log.param_grid['C'], grid_log.cv_results_['mean_test_score'][1::4], label='train', marker='.')
grid_log.fit(X_test, y_test)
plt.plot(grid_log.param_grid['C'], grid_log.cv_results_['mean_test_score'][1::4], label='test', marker='.')
plt.title("AUC dependency params")
plt.xlabel('C')
plt.ylabel('AUC')
plt.legend(loc="lower right")
plt.show()
```
